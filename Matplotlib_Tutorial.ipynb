# Line Plot and Basics

# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['Google' , 'Amazon' , 'Microsoft' , 'Facebook']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
plt.bar(company,revenue)
```

```python
plt.bar(company,revenue)
plt.bar(company,profit)
```

```python
plt.bar(company,revenue , width = 0.4)
plt.bar(company,profit , width = 0.4)
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue)
```

```python
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
```

```python
plt.ylabel("Revenue(Billion $)")
plt.title("Us Tech Stocks")
plt.xticks(xpos,company)
plt.bar(xpos,revenue , label = "Revenue")

plt.legend()
```

```python
plt.ylabel("Revenue(Billion $)")
plt.title("Us Tech Stocks")
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue , label = "Revenue")
plt.bar(xpos+0.2,profit , label = "Profit")

plt.legend()
```

```python
plt.ylabel("Revenue(Billion $)")
plt.title("Us Tech Stocks")
plt.xticks(xpos,company)
plt.bar(xpos-0.2,revenue, width = 0.4 , label = "Revenue")
plt.bar(xpos+0.2,profit, width = 0.4 , label = "Profit")

plt.legend()
```

```python
plt.ylabel("Revenue(Billion $)")
plt.title("Us Tech Stocks")
plt.yticks(xpos,company)
plt.barh(xpos-0.2,revenue, label = "Revenue")
plt.barh(xpos+0.2,profit, label = "Profit")

plt.legend()
```

# Histograms

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar , bins = 3)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar , bins = 3 , rwidth = 0.95)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar , bins = [80,100,125,150] , rwidth = 0.95)
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar , bins = [80,100,125,150] , rwidth = 0.95 , color = 'g')
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar , bins = [80,100,125,150] , rwidth = 0.95 , color = 'g' , histtype = 'step')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men , blood_sugar_women] , bins = [80,100,125,150] , color = ['red','green'])
```

```python
plt.hist([blood_sugar_men , blood_sugar_women] , bins = [80,100,125,150] , color = ['red','green'] , rwidth = 0.95)
```

```python
plt.hist([blood_sugar_men , blood_sugar_women] , bins = [80,100,125,150] , color = ['red','green'] , rwidth = 0.95 ,label = ['men' , 'woman'])
plt.legend()
```

```python
plt.hist([blood_sugar_men , blood_sugar_women] , bins = [80,100,125,150] , color = ['red','green'] , rwidth = 0.95 ,label = ['men' , 'woman'])
plt.legend()
plt.xlabel('sugar range')
plt.ylabel('Total Number of patients')
plt.title('Blood Sugar Analysis')
```

```python
plt.hist([blood_sugar_men , blood_sugar_women] , bins = [80,100,125,150] , color = ['red','green'] , rwidth = 0.95 ,label = ['men' , 'woman'],
         orientation = 'horizontal')
plt.legend()
plt.ylabel('sugar range')
plt.xlabel('Total Number of patients')
plt.title('Blood Sugar Analysis')
```

# Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent" , "Food" , "Phone/Internet Bill" , "Car" , "Other Utilities"]
```

```python
plt.pie(exp_vals , labels = exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5,autopct = '%f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5,autopct = '%0.2f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5,autopct = '%0.0f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5,autopct = '%0.1f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5,autopct = '%0.2f%%' , shadow = True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5,autopct = '%0.2f%%' , shadow = True , explode =[0,0,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5,autopct = '%0.2f%%' , shadow = True , explode =[0,0.2,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5,autopct = '%0.2f%%' , shadow = True , explode =[0,0.5,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5,autopct = '%0.2f%%' , shadow = True , explode =[0,0.2,0.5,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals , labels = exp_labels , radius = 1.5,autopct = '%0.2f%%' , shadow = True , explode =[0,0.1,0.1,0,0] , startangle = 180)
plt.show()
```
